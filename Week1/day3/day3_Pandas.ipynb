```python
import pandas as pd
```

```python
df= pd.read_csv("/content/Titanic-Dataset.csv")
```

```python
# Q1: How many passengers were in First Class?
ans1 = df[df['Pclass'] == 1].shape[0]
ans1
```

```python
# Q2 How many passenger were male?
ans2 = df[df['Sex'] == 'male'].shape[0]
ans2
```

```python
# Q3 how many passenger were belong to the cabinet c85
ans3 = df[df['Cabin'] == 'C85'].shape[0]
ans3
```

```python
# Q4: Get the list of all female passengers who survived.
ans4 = df[(df['Sex'] == 'female') & (df['Survived']== 1) ]['Name']
ans4
```

```python
# Q5: What is the average age of passengers who did not survive?
ans5 = df[(df['Survived'] == 0)]['Age'].mean()
ans5
```

```python
# Q6 List a passenger name who paid a fare greater than $200
ans6 = df[(df['Fare'] > 200)]['Name']
ans6
```

```python
# Q7 How many passenger were embarked from S
ans7 = df[df['Embarked'] == 'S'].shape[0]
ans7
```

```python
# Q8 how many passenger had age greater than 60 and survived
ans8 = df[(df['Age']  > 60 )  & (df['Survived'] == 1)].shape[0]
ans8
```

```python
# Q9 list the detail of passenger who have no siblings and no parents on board
ans9 = df[(df['SibSp'] == 0) & ( df['Parch'] == 0)]
ans9
```

```python
# Q10 who was the oldest passenger on boat
ans10 = df.loc[df['Age'].idxmax()]
ans10
```
