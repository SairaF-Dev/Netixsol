**Importing necessary libraries**

```python
import pandas as pd
```

**Data Loading**

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

**Identifying the missing Values**

```python
missing_count = df.isnull().sum()
missing_count
```

**Handling Missing Values**

Cabin has too many missing values, so we drop it

```python
df = df.drop(columns=['Cabin'])
```

```python
# df.head(10)
```

 Age has 177 missing values so fill them with mean to preserve data


```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
missing_count = df.Age.isnull().sum()
missing_count
```

**Fix Wrong DataType**

Although the age column was filled but it retained a float datatype due to previous NAN values.

```python
df.Age.dtype
```

```python
# Convert the age column to integer type
df['Age'] = df['Age'].astype(int)
```

**Remove Duplicates**

```python
total_duplicates = df.duplicated().sum()
print(total_duplicates)
```

 No duplicates fount, so we dont need to run drop_duplicates()

**Create 2 New Derived Columns**

```python
# Combineing SibSp and Parsch helps analyze the total family size
df['FamilySize'] = df['SibSp']+df['Parch']+1
```

```python
# Categorizing age helps simplifying analysis.
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x<18 else 'Adult' )
```

**Reshape Data (Summary Table)**

Aggreagated data by 'Pclass' to analyze the count of passengers and their average age, highlighting demographic patterns across class

```python
# Creating a summary table to see survival rates by class
summary_table = df.groupby('Pclass').agg({'Survived': 'count', 'Age':'mean',})
summary_table
```

**Mapping and Merging**

Used a left join to map numerical 'Pclass' values to descriptive labels ('First', 'Second', 'Third') while preserving all original dat

```python
#  Create mapping
class_mapping = pd.DataFrame({'Pclass': [1,2,3], 'ClassName': ['First', 'Second', 'Third']})

# Merge to add the label

df = df.merge(class_mapping, on='Pclass', how = 'left')

```
