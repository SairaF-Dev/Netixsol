```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
titanic_data_file_path = '/content/Titanic-Dataset.csv'
```

```python
df = pd.read_csv(titanic_data_file_path)
```

```python
df.head(2)
```

**Histogram**

This histogram shows the 'Passenger Age' distribution. The graph clearly shows that most passengers between the ages of 20 and 30 with a significant peak in third class. The KDE curve demonstrate how the age distribution trends shift across classes.

```python
# Histogram: Age distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age', kde=True, hue= 'ClassName')
plt.title('Distribution of Passenger Age by Class')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Bar Chart**

This bar chart illustrate the average fair paid by passenger across three different classes. Where 'First' class 'average' fare is higher compared to 'Second' and 'Third' classes. The error bars on the top represent the actual fare in that class were spread further away from the average fare, while a shorter line means the fare were more concistent and closer to the average



```python
# Bar Chart: Average Fare by ClassName
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='ClassName', y='Fare', hue='ClassName', legend=True)
plt.title('Average Fare by Class')
plt.xlabel('Class')
plt.ylabel('Average Fare')
plt.show()
```

**BoxPlot**

The boxplot visualizes the 'Fare' distribution across different passenger classes. it highlights the median price for each class, the range of fare (the box area), and identifies outliers (the circles) which are unusually high fares that fall outside the typical range of each class.

```python
# Boxplot: Fare distribution by ClassName
plt.figure(figsize=(8,5))
sns.boxplot(data= df, x='ClassName', y='Fare', hue='ClassName', legend= True)
plt.title('Fare Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Fare')
plt.show()
```

**Heatmap**

This heatmap displays the correlation coefficients between numeric variables. Darker red cells indicate a strong postive corelation (e.g SibSp anf FamilySize at 0.89), while darker blue cells indicate negative correlation(e.g, Pclass and Fare at -0.55), revealing how different features influence each other within the dataset.

```python
# Corelation Heatmap
plt.figure(figsize=(10,8))
numberic_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
